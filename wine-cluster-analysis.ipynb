# Clustering with the metod K-means

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
%matplotlib inline
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
```

# Import data

```python
wine = pd.read_csv('../input/wine-dataset-for-clustering/wine-clustering.csv')
```

```python
wine.info()
```

```python
wine_norm=(wine-wine.min())/(wine.max()-wine.min())
wine_norm.head()
```

```python
wine_norm.describe()
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i,max_iter=300)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clustering =KMeans(n_clusters =3,max_iter=300)
clustering.fit(wine_norm)
```

```python
wine['kMeans_Clusters']= clustering.labels_
wine.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_wine=pca.fit_transform(wine_norm)
pca_wine_df = pd.DataFrame(data = pca_wine, columns=['Component_1','Component_2'])
pca_wine_name = pd.concat([pca_wine_df,wine[['kMeans_Clusters']]],axis=1)
pca_wine_name
```

```python
fig =plt.figure(figsize = (6,6))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Component 1',fontsize = 15)
ax.set_ylabel('Component 2',fontsize = 15)
ax.set_title('Main components',fontsize=20)
color_theme = np.array(['blue','green','orange'])
ax.scatter(x= pca_wine_name.Component_1, y=pca_wine_name.Component_2,
          c= color_theme[pca_wine_name.kMeans_Clusters],s=50)
plt.show
```
